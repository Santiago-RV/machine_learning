# Let's start on the basics. Variables and arrays

### First, import some libraries

```
import numpy as np
```

```
array_1 = np.array([5,9,4,7,8,6,8,6]) # Create an array of N values
array_2 = np.random.randint(1, 50, 8) # Create an array of random int values
            # with the same lenght as array_1

print(array_1,array_2) # print both array_1 and array_2
print(array_1.shape, array_2.shape) # print both array_1 and array_2 shapes
print(array_1 + array_2) # print summation of array_1 and array_2
```

```
array_1[3]=70 # Replace the value of array_1 in pos 3 by 70
print(array_1 + array_2) # Print summation again
```

```
array_3=array_1 * array_2 # Apply element-wise multiplication among array_1 and array_2

print(array_3) # Print results
```

```
# Playing with image intensities
img_gray = cv2.imread('/content/drive/MyDrive/LP1/machine_learning/Images/lena.jpg')

for r in range(img_gray.shape[0])  :  # Go through rows
  for c in range(img_gray.shape[1]) : # Go through cols
   if np.any(img_gray[r, c] > 150): # Create a simple condition
    img_gray[r, c] = 255


plt.imshow(img_gray, cmap="gray" )
```

### Crop the image - taking regions of interest (ROI)

```
# Extract some ROI
x1, y1, x2, y2 = 200, 200, 400, 400  # Reemplaza con las coordenadas de tu ROI
roi = img[y1:y2, x1:x2]

# Mostrar la ROI
plt.imshow(roi)
plt.axis('off')
plt.show()
```

```
#img_gray[ _ , _ ] = roi
#plt.imshow(img_gray, _ )

# Convierta la ROI a escala de grises
roi_gray = cv2.cvtColor(roi, cv2.COLOR_RGB2GRAY)

# Asigne la ROI a la imagen en escala de grises
img_gray[y1:y2, x1:x2] = roi_gray

# Mostrar la imagen en escala de grises
plt.imshow(img_gray, cmap='gray')  # Utiliza cmap='gray' para mostrar la imagen en escala de grises
plt.axis('off')
plt.show()
```
