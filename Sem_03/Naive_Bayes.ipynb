```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import load_iris # Iris dataset
```

```
iris = load_iris() # Load the iris dataset
iris_x = iris.data[:,:2] # Features. We'll pick 2
iris_y = iris.target # Labels

feature_names=iris.feature_names
# Print shapes
print ( iris_x.shape, iris_y.shape, feature_names )
_, ax=plt.subplots(figsize=(6,5))
ax.scatter(iris_x[:,0], iris_x[:,1], c=iris_y, cmap='gnuplot') #Columna 1 del ancho Nro 1 y columna 2 de altura Nro1, c=etiquetas de los datos
ax.set_xlabel(feature_names[0]) #Nombres de eje horizontal
ax.set_ylabel(feature_names[1]) #Nombres de eje vertical
```

```
# Instance kNN object
naive_bayes = GaussianNB()

# Training
naive_bayes.fit( iris_x , iris_y ) #(num_samples, num_features), (num_samples)
```

```
## Plot decision boundaries
offset=4
h_min, h_max =iris_x[:,0].min()-offset,iris_x[:,0].max()+offset  #MINIMOS Y MAXIMOS DEL EJE HORIZONTAL
v_min, v_max =iris_x[:,1].min()-offset,iris_x[:,1].max()+offset  #MINIMOS Y MAXIMOS DEL EJE VERTICAL
h_grid,v_grid=np.meshgrid(np.arange(h_min,h_max, 0.03), np.arange(v_min,v_max,0.03))

#print(h_grid.shape,v_grid.shape,h_grid.reshape(-1).shape,v_grid.reshape(-1).shape)
#print(np.c_[h_grid.reshape(-1),v_grid.reshape(-1)].shape)
#print(pred_grid.shape)
#pred_grid= pred_grid.reshape(h_grid.shape)

print(pred_grid.shape, h_grid.shape, h_grid.ravel().shape,v_grid.reshape(-1).shape)
print(np.c_[h_grid.ravel(),v_grid.reshape(-1)].shape)
```

```
pred_grid=naive_bayes.predict(np.c_[h_grid.ravel(),v_grid.reshape(-1)])
pred_grid = pred_grid.reshape(h_grid.shape)
print(pred_grid.shape,h_grid.shape,v_grid.shape)
```

```
__,ax =plt.subplots(figsize=(6,5))
ax.pcolormesh(h_grid,v_grid,pred_grid,cmap='terrain') #colormesh pinta la cuadricula
ax.scatter(iris_x[:,0],iris_x[:,1],c=iris_y, edgecolors='k',cmap='gnuplot')

ax.set_xlabel(feature_names[0]) #Nombres de eje horizontal
ax.set_ylabel(feature_names[1]) #Nombres de eje vertical
```
